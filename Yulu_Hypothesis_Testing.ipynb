## analysing numeric columns

```python
# these are the numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
num_cols
```

```python
# histogram and dist plot  for numeric columns
plt.figure(figsize=(16, 12))

for i, col in enumerate(num_cols, 1):
    plt.subplot(4, 3, i)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

```python
#  these are the categorical columns
cat_cols = ['season', 'holiday', 'workingday', 'weather']
#count plot for categorical columns
plt.figure(figsize=(14, 10))

for i, col in enumerate(cat_cols, 1):
    plt.subplot(2, 2, i)
    sns.countplot(data=df, x=col)
    plt.title(f'Countplot of {col}')

plt.tight_layout()
plt.show()
```

```python
# this is the pie chart for the categorical columns
plt.figure(figsize=(14, 10))

for i, col in enumerate(cat_cols, 1):
    plt.subplot(2, 2, i)
    df[col].value_counts().plot(
        kind='pie',
        autopct='%1.1f%%',
        startangle=90
    )
    plt.title(f'Pie Chart of {col}')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
## now detecting and haldeling outliers

# using box plot to detect outliers
plt.figure(figsize=(16, 12))

for i, col in enumerate(num_cols, 1):
    plt.subplot(4, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```

```python
# now cliping outlires usig IQR

df_clipped = df.copy()

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df_clipped[col] = df[col].clip(lower=lower, upper=upper)
```

```python
# now lets check if we have treated the outlires or not
plt.figure(figsize=(16, 12))

for i, col in enumerate(num_cols, 1):
    plt.subplot(4, 3, i)
    sns.boxplot(y=df_clipped[col])
    plt.title(f'Boxplot of {col} (After Clipping)')

plt.tight_layout()
plt.show()
# every thing loooks fine
```

## 2. Try establishing a Relationship between the Dependent and Independent Variables.

```python
#lets plot a correaltion heatmap
num_df = df.select_dtypes(include=['int64', 'float64'])
num_df.columns
corr_matrix = num_df.corr()
```

```python
# here is the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    linewidths=0.5
)
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
# now lets check for the highly correlated value and remove them
high_corr_pairs = (
    corr_matrix.abs()
    .where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    .stack()
    .sort_values(ascending=False)
)

high_corr_pairs[high_corr_pairs > 0.85]
```

```python
# lets remove the above columns
df_reduced = df.drop(columns=['temp','registered'])
```

```python
# now lets verify correlation 
corr_matrix_reduced = df_reduced.select_dtypes(include=['int64', 'float64']).corr()

plt.figure(figsize=(12, 8))
sns.heatmap(
    corr_matrix_reduced,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)
plt.title('Correlation Heatmap After Removing Highly Correlated Variables')
plt.show()
```
